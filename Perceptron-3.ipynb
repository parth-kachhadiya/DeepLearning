```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X, Y = make_classification(n_samples=100, n_features=2, n_informative=1,
                           n_redundant=0, n_classes=2, n_clusters_per_class=1,
                           random_state=41, hypercube=False, class_sep=15)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:,0], X[:,1], c=Y, cmap='winter', s=100)
plt.show()
```

```python
X.shape, Y.shape
```

```python
def perceptron(x, y, learning_rate=0.1, epochs=1000):
    w1 = w2 = b = 1

    for j in range(epochs):
        for i in range(x.shape[0]):
            z = w1 * x[i][0] + w2 * X[i][1] + b
    
            if z * y[i] < 0:
                w1 = w1 + learning_rate * y[i] * x[i][0]
                w2 = w2 + learning_rate * y[i] * x[i][1]
                b = b + learning_rate * y[i]

    return w1, w2, b
```

```python
w1, w2, b = perceptron(X, Y)
```

```python
m = -(w1/w2)
c = -(b/w2)
m, c
```

```python
x_inp = np.linspace(-3, 3, 100)
y_inp = m * x_inp + c

plt.figure(figsize=(10, 6))
plt.plot(x_inp, y_inp, color='red', linewidth=3)
plt.scatter(X[:,0], X[:,1], c=Y, cmap='winter', s=100)
plt.ylim(-3, 2)
plt.show()
```
