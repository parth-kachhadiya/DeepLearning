# Problems in perceptrons

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.linear_model import Perceptron
from mlxtend.plotting import plot_decision_regions
```

```python
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```python
or_data['in1'] = [1, 1, 0, 0]
or_data['in2'] = [1, 0, 1, 0]
or_data['out'] = [1, 1, 1, 0]
```

```python
and_data['in1'] = [1, 1, 0, 0]
and_data['in2'] = [1, 0, 1, 0]
and_data['out'] = [1, 0, 0, 0]
```

```python
xor_data['in1'] = [1, 1, 0, 0]
xor_data['in2'] = [1, 0, 1, 0]
xor_data['out'] = [0, 1, 1, 0]
```

```python
print(and_data)
sns.scatterplot(data=and_data, x='in1', y='in2', hue='out', s=200)
```

```python
print(or_data)
sns.scatterplot(data=or_data, x='in1', y='in2', hue='out', s=200)
```

```python
print(xor_data)
sns.scatterplot(data=xor_data, x='in1', y='in2', hue='out', s=200)
```

```python
clf_1 = Perceptron()
clf_2 = Perceptron()
clf_3 = Perceptron()
```

```python
clf_1.fit(and_data.iloc[:,:-1].values, and_data.iloc[:,-1].values)
clf_2.fit(or_data.iloc[:,:-1].values, or_data.iloc[:,-1].values)
clf_3.fit(xor_data.iloc[:,:-1].values, xor_data.iloc[:,-1].values)
```

```python
clf_1.coef_, clf_1.intercept_
```

```python
x = np.linspace(-1, 1, 5)
y = -x + 1
```

```python
plt.plot(x, y)
sns.scatterplot(data=and_data, x='in1', y='in2', hue='out', s=200)
```

```python
x1 = np.linspace(-1, 1, 5)
y1 = -x1 + 0.5
```

```python
clf_2.coef_, clf_2.intercept_
```

```python
plt.plot(x1, y1)
sns.scatterplot(data=or_data, x='in1', y='in2', hue='out', s=200)
```

```python
clf_3.coef_, clf_3.intercept_
```

```python
plot_decision_regions(xor_data.iloc[:,:-1].values, xor_data.iloc[:,-1].values, clf=clf_3, legend=2)
```
